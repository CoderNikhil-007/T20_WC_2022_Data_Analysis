```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
```

```python
df_match.rename({'scorecard' : 'id'}, axis = 1,inplace = True)
```

```python
df_match.head()
```

```python
match_id_dict = {}
for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + 'Vs' + row['team1']
    
    match_id_dict[key1] = row['id']
    match_id_dict[key2] = row['id']

match_id_dict
```

**Batting Summary**

```python
with open('t20_json_files/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.tail()
```

```python
df_batting['Out/NotOut'] =  df_batting.dismissal.apply(lambda x:'Out' if len(x)>0 else 'NotOut')
df_batting.tail()
```

```python
df_batting.drop(columns = 'dismissal', inplace = True)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(11)
```

```python
df_batting['id'] = df_batting['match'].map(match_id_dict)

df_batting.head(11)
```

```python
with open('t20_json_files/t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])

df_bowling = pd.DataFrame(all_records)
df_bowling.tail()
```

```python
df_bowling['id'] = df_bowling['match'].map(match_id_dict)

df_bowling.head(11)
```
